# Mistral NeMo 12B Archiver

This notebook downloads the required files for the Mistral NeMo 12B model from Hugging Face, explicitly omitting the redundant 24.5 GB consolidated weight file. 

It then compresses the necessary files into a high-compression zip archive and saves it directly to your designated Google Drive folder.

```python
# ==========================================
# USER CONFIGURATION
# ==========================================

# Paste your Hugging Face access token here
HF_TOKEN = ""

# Target Google Drive output directory
GDRIVE_OUTPUT_DIR = "/content/drive/MyDrive/soulclone"

# The final zip file name
ZIP_FILENAME = "Mistral-Nemo-Instruct-2407.zip"
```

```python
import os
from google.colab import drive

print("Mounting Google Drive...")
drive.mount('/content/drive')

print("\nInstalling required libraries...")
!pip install -q huggingface_hub

# Create the target directory on your Drive if it does not exist
os.makedirs(GDRIVE_OUTPUT_DIR, exist_ok=True)
print(f"\nOutput directory verified: {GDRIVE_OUTPUT_DIR}")
```

```python
from huggingface_hub import snapshot_download, login

if HF_TOKEN:
    print("Authenticating with Hugging Face...")
    login(token=HF_TOKEN)
else:
    print("WARNING: No HF_TOKEN provided. Download will likely fail.")

REPO_ID = "mistralai/Mistral-Nemo-Instruct-2407"
LOCAL_DOWNLOAD_DIR = "/content/Mistral-Nemo-Local"

# The massive redundant file and unnecessary text files
IGNORE_FILES = ["consolidated.safetensors", "*.md", ".gitattributes"]

print(f"\nStarting high-speed download for: {REPO_ID}")
print("Pulling directly into temporary Colab storage...")

snapshot_download(
    repo_id=REPO_ID,
    local_dir=LOCAL_DOWNLOAD_DIR,
    ignore_patterns=IGNORE_FILES
)

print("\nDownload complete!")
```

```python
import zipfile
import time

final_zip_path = os.path.join(GDRIVE_OUTPUT_DIR, ZIP_FILENAME)

print(f"Compressing downloaded files into {ZIP_FILENAME}...")
print("Using maximum DEFLATED compression. This will take several minutes...")

start_time = time.time()

# Create zip with maximum compression (level 9)
with zipfile.ZipFile(final_zip_path, 'w', zipfile.ZIP_DEFLATED, compresslevel=9) as zipf:
    for root, dirs, files in os.walk(LOCAL_DOWNLOAD_DIR):
        for file in files:
            file_path = os.path.join(root, file)
            # Strip the absolute path so the zip extracts cleanly
            arcname = os.path.relpath(file_path, start=LOCAL_DOWNLOAD_DIR)
            zipf.write(file_path, arcname)

duration = time.time() - start_time
print(f"\nSuccess! Archive saved to: {final_zip_path}")
print(f"Compression took {duration:.2f} seconds.")
print("You can now safely close this notebook and use the zip in your main fine-tuning script.")
```
